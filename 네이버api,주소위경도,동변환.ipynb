```python
import urllib
import json
import pandas as pd
```

```python
def search_map(search_text):
    client_id = 'id'
    client_secret = 'secret'
    encText = urllib.parse.quote(search_text)
    url = 'https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query='+encText
    request = urllib.request.Request(url)
    request.add_header('X-NCP-APIGW-API-KEY-ID','') #####아이디
    request.add_header('X-NCP-APIGW-API-KEY','') ###키
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
#         print(response_body.decode('utf-8'))
        return response_body.decode('utf-8')
    else:
        print("Error Code:" + rescode)
```

```python
def make_location(geo_list):
    x = []
    y = []
    for geo_location in geo_list:
        temp_map = search_map(geo_location)
        temp_map = json.loads(temp_map)
        try:
            temp_map = temp_map['addresses'][0]
            x.append(float(temp_map['x']))
            y.append(float(temp_map['y']))
        except IndexError:
            pass
    return x, y
```

```python
def make_pandas_location(geo_pandas):
    
    geopd = pd.DataFrame(columns=('geo_location', 'x', 'y'))
    j=0
    for geo_location in geo_pandas['Full Address']:
        temp_map = search_map(geo_location)
        temp_map = json.loads(temp_map)
        try:
            temp_map = temp_map['addresses'][0]
            dong = temp_map['addressElements'][2]['longName'] #### 동면가져오는 코드 수정
            geopd = geopd.append({'geo_location':geo_location,
                                  'dong':dong, ### 동추가
                                 'x':float(temp_map['x']),
                                 'y':float(temp_map['y'])
                                    }, ignore_index=True )
            
            j+=1
            if j%1000 == 0:
                print (j/1000,'/','20 번째')
        except IndexError:
            pass
    return geopd
```
