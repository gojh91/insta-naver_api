```python
import requests
import json
import urllib.request
import datetime as dt
import pandas as pd
import csv
```

```python
class insta:
    def __init__(self, keyword):
        self.keyword = keyword
        
    def toCSV(self, list_a):
        print("toCSV")
        with open( self.keyword + '.csv', 'a', encoding='utf-8', newline='') as file :
            fieldnames = ['text', 'date', 'like_count', 'explain', 'each_url']
            csvfile = csv.DictWriter(file, fieldnames=fieldnames)
            for row in list_a:
                csvfile.writerow(row)    
    
    def makeCSV(self):
        with open(self.keyword + '.csv', 'w', encoding='utf-8', newline='') as file:
            fieldnames = ['text', 'date', 'like_count', 'explain', 'each_url']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
    
    
    def insta_crawling(self):
        print("insta_crawling")
        
        after = ""
        i = 0
        self.makeCSV()

        while True:
            list_a = []
            try:
                url = 'https://www.instagram.com/graphql/query/?query_hash=174a5243287c5f3a7de741089750ab3b&variables={"tag_name":"' + self.keyword + '","first": 50 ,"after": "' + after + '"}'  # 요청할 주소
                resp = requests.get(url)
                json_data = json.loads(resp.text)

#                 print(url)
                after = json_data['data']['hashtag']['edge_hashtag_to_media']['page_info']['end_cursor']
                for edge in json_data['data']['hashtag']['edge_hashtag_to_media']['edges'] :

                    dict_a = {}

                    print('i : ' + str(i))

                    text = edge['node']['edge_media_to_caption']['edges'][0]['node']['text']
                    dict_a['text'] = text
#                     print('text : ' + text)

                    timestamp = edge['node']['taken_at_timestamp']
                    date = dt.datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S')
                    dict_a['date'] = date
#                     print('date : ' + date)

                    shortcode = edge['node']['shortcode']
                    each_url = 'https://www.instagram.com/graphql/query/?query_hash=477b65a610463740ccdb83135b2014db&variables={"shortcode":"' + shortcode + '"}'
                    dict_a['each_url'] = each_url
#                     print('each_url : ' + each_url)

                    ###숫자 형식으로 가져옴
                    like_count = edge['node']['edge_liked_by']['count']
                    dict_a['like_count'] = like_count
#                     print('like_count : ' + str(like_count))


                    explain = edge['node']['accessibility_caption']
                    dict_a['explain'] = explain
#                     print('explain : ' + explain)

#                     each_resp = requests.get(each_url)
#                     each_json_data = json.loads(each_resp.text)
#                     try:
#                         each_location = each_json_data['data']['shortcode_media']['location']['name']
#                         print('each_location : ' + each_location)
#                         address_json = each_json_data['data']['shortcode_media']['location']['address_json']
#                         print('address_json : ' + address_json)
#                     except:
#                         pass

                    i += 1

                    list_a.append(dict_a)
                    
                    

#                     writer.writerow(dict_a) 

#                     print("\n\n\n\n")
            except:
#                 print(list_a)
                self.toCSV(list_a)
```

```python
class insta_each:
    def __init__(self, csvname):
        self.csvname = csvname
        self.list_a = []
    def toCSV(self):
        print("toCSV")
        with open( self.csvname + '_each.csv', 'a', encoding='utf-8', newline='') as file :
            fieldnames = ['each_url', 'each_location', 'address_json']
            csvfile = csv.DictWriter(file, fieldnames=fieldnames)
            for row in self.list_a:
                csvfile.writerow(row)    

    def makeCSV(self):
        print("makeCSV")
        with open(self.csvname + '_each.csv', 'w', encoding='utf-8', newline='') as file:
            fieldnames = ['each_url', 'each_location', 'address_json']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            
    def insta_crawling_each(self):
        i = 0
        self.makeCSV()
        df = pd.read_csv(self.csvname + '.csv')
        print('1####')
        for each_url in df['each_url']:
            print(i)
            i += 1
            print(each_url)
            dict_a = {}
            try:
                each_resp = requests.get(each_url)
                each_json_data = json.loads(each_resp.text)
                print('2####')

                dict_a['each_url'] = each_url
                each_location = each_json_data['data']['shortcode_media']['location']['name']
                dict_a['each_location'] = each_location
                address_json = each_json_data['data']['shortcode_media']['location']['address_json']
                dict_a['address_json'] = address_json
                self.list_a.append(dict_a)
                print('3####')
            except Exception as e:
                print(e)
                
            

```
